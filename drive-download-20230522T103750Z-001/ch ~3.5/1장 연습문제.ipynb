```python
import torch
```

```python
# 1. 2D텐서를 만들고 차원0 위치에 크기가 1인 차원을 추가하시오.
x = torch.Tensor(3,3)
x = x.unsqueeze(0)
x
```

```python
# 2. 이전 텐서에 추가한 차원을 삭제하시오.
x = x.squeeze(0)
x
```

```python
# 3. 범위가 3-7 이고 크기가 5*3인 랜덤한 텐서를 만드시오.
y = torch.rand(5,3) * 4 + 3
y
```

```python
# 4. 정규분포를 사용해 텐서를 만드시오.
z = torch.rand(3,3)
z.normal_() # _를 붙이면 해당 객체에 바로 적용된다.
```

```python
# 5. 텐서 torch.Tensor([1,1,1,0,1])에서 0이 아닌 원소의 인덱스를 추출하시오.
a = torch.Tensor([1,1,1,0,1])
torch.nonzero(a) # 2차원 텐서 반환 / 갯수는 count_nozero()함수 사용
```

```python
# 6. 크기가 3*1인 랜덤한 텐서를 만들고 네 벌을 복사해 쌓으시오.
a = torch.rand(3,1)
a.expand(3,4) # expand는 크기가 1인 차원을 변경하므로 1인 아닌 차원은 (-1,1)이런식으로 -1로 나타내도 된다.
```

```python
# 7. 2차원 행렬 두 개의 배치행렬곱셈을 계산하시오
a = torch.rand(3, 4, 5)
b = torch.rand(3, 5, 4)
print(a)
print(b)
torch.bmm(a,b) # (a,b,c) 에서 a를 배치차원으로 인식 따라서 a는 둘 모두 같아야 한다.
```

```python
# 8. 3차원 행렬과 2차원 행렬의 배치행렬곱셈을 계산하시오.
a = torch.rand(3,4,5)
b = torch.rand(5,4)
torch.matmul(a,b)
'''
좀 더 풀어써보면 
a 와 b의 배치 차원을 맞춰 주기 위해 b.squeeze(0)으로 (1,5,4)의 형태로 만들어주고 
다시 b.expand(3,-1,-1)을 해준다. 이를 요약하면 b.squeeze(0).expand(3,-1,-1)이 된다.
또는 파이썬의 *,**을 이용해서(파라미터로 몇개를 받을지 모르는 경우에 사용)
expand(a.size(0),*b.size())를 사용한다.
'''
```
